```python
from selenium import webdriver 
import time as TIME
import pandas as pd
import datetime 


import warnings
warnings.filterwarnings("ignore")
```

```python
def get_data(url):
    data_element = browser.find_elements_by_xpath(url)
    datas = [x.text for x in data_element]
    return datas
```

```python
option = webdriver.ChromeOptions()

option.add_argument("â€” incognito")

executable_path = "/Users/cuong/Documents/Selenium/chromedriver"
#executable_path = "C:\\Users\\truon\\Downloads\\chromedriver_win32\\chromedriver"

browser = webdriver.Chrome(executable_path = executable_path, 
                            chrome_options=option)

browser.get("https://trade-hn.vndirect.com.vn/chung-khoan/phai-sinh")


TIME.sleep(5)


```

```python
columns = ["Time","Price", "Change","Volume","CumVol"]

df = pd.DataFrame(columns = columns)

start_time = TIME.time()

date = get_data("//*[@id='clock-date']")

#browser.find_element_by_xpath("//*[@id='banggia-phaisinh-body']/tr[2]/td[1]").click()

#TIME.sleep(2)
i = 1

while True:
    try:
        
        time = get_data("//*[@id='derivative-info-body-VN30F1810']/div/div[" + str(i) + "]/span[1]")
 
        price = get_data("//*[@id='derivative-info-body-VN30F1810']/div/div[" + str(i) + "]/span[2]")
        change = get_data("//*[@id='derivative-info-body-VN30F1810']/div/div[" + str(i) + "]/span[3]")
        vol = get_data("//*[@id='derivative-info-body-VN30F1810']/div/div[" + str(i) + "]/span[4]")
        cumvol = get_data("//*[@id='derivative-info-body-VN30F1810']/div/div[" + str(i) + "]/span[5]")
        
        if change == ['']:
            change = ['0']

            
        df.loc[i-1] = (time[0], price[0], change[0], vol[0], cumvol[0])
        #print(i, time, price, change, vol, cumvol)
        i += 1
        
    except:
        print("                       Mission Accomplished")
        print("--- %s seconds ---" % (TIME.time() - start_time))
        break

df = df.iloc[::-1].reset_index(drop = True)

```
